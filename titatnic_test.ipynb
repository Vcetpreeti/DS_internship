```python
#4. Decision Tree

import pandas as pd
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Assuming your DataFrame is named df and is already loaded

# Drop unnecessary columns and prepare feature matrix x and target variable y
x = df.drop(["PassengerId", "Name", "Cabin", "Ticket", "Parch", "SibSp"], axis=1)
y = df["Survived"]

# Map categorical variables to numerical values
x['Sex'] = x['Sex'].map({'female': 0, 'male': 1})
x['Embarked'] = x['Embarked'].map({'Q': 0, 'S': 1, 'C': 2})

# Handle missing values if any
x = x.fillna(x.mean())

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=0.3)

# Initialize and train the Decision Tree Classifier
dt = tree.DecisionTreeClassifier()
dt.fit(X_train, y_train)

# Predict the test set
y_pred = dt.predict(X_test)

# Print the accuracy score
print(accuracy_score(y_test, y_pred))
```
